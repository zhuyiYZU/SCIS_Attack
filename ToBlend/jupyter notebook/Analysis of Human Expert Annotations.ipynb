```python
# import warnings
# warnings.filterwarnings('error', category=DeprecationWarning)
import pandas as pd
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

# Import data and process

```python
df = pd.read_csv('datasets/annotation/Annotation_all.csv', encoding='latin-1')
df
```

```python
for i in tqdm(range(len(df))):
    df.loc[i,'coh-a'] = float((df.loc[i,'coh-a-1']+df.loc[i,'coh-a-2']+df.loc[i,'coh-a-3'])/3)
    df.loc[i,'flu-a'] = float((df.loc[i,'flu-a-1']+df.loc[i,'flu-a-2']+df.loc[i,'flu-a-3'])/3)
    
    df.loc[i,'coh-b'] = float((df.loc[i,'coh-b-1']+df.loc[i,'coh-b-2']+df.loc[i,'coh-b-3'])/3)
    df.loc[i,'flu-b'] = float((df.loc[i,'flu-b-1']+df.loc[i,'flu-b-2']+df.loc[i,'flu-b-3'])/3)
    
    df.loc[i,'coh-c'] = float((df.loc[i,'coh-c-1']+df.loc[i,'coh-c-2']+df.loc[i,'coh-c-3'])/3)
    df.loc[i,'flu-c'] = float((df.loc[i,'flu-c-1']+df.loc[i,'flu-c-2']+df.loc[i,'flu-c-3'])/3)
    
    df.loc[i,'coh-d'] = float((df.loc[i,'coh-d-1']+df.loc[i,'coh-d-2']+df.loc[i,'coh-d-3'])/3)
    df.loc[i,'flu-d'] = float((df.loc[i,'flu-d-1']+df.loc[i,'flu-d-2']+df.loc[i,'flu-d-3'])/3)
    
    df.loc[i,'coh-e'] = float((df.loc[i,'coh-e-1']+df.loc[i,'coh-e-2']+df.loc[i,'coh-e-3'])/3)
    df.loc[i,'flu-e'] = float((df.loc[i,'flu-e-1']+df.loc[i,'flu-e-2']+df.loc[i,'flu-e-3'])/3)
    
    df.loc[i,'coh-f'] = float((df.loc[i,'coh-f-1']+df.loc[i,'coh-f-2']+df.loc[i,'coh-f-3'])/3)
    df.loc[i,'flu-f'] = float((df.loc[i,'flu-f-1']+df.loc[i,'flu-f-2']+df.loc[i,'flu-f-3'])/3)
    
    df.loc[i,'coh-g'] = float((df.loc[i,'coh-g-1']+df.loc[i,'coh-g-2']+df.loc[i,'coh-g-3'])/3)
    df.loc[i,'flu-g'] = float((df.loc[i,'flu-g-1']+df.loc[i,'flu-g-2']+df.loc[i,'flu-g-3'])/3)
    
    df.loc[i,'coh-h'] = float((df.loc[i,'coh-h-1']+df.loc[i,'coh-h-2']+df.loc[i,'coh-h-3'])/3)
    df.loc[i,'flu-h'] = float((df.loc[i,'flu-h-1']+df.loc[i,'flu-h-2']+df.loc[i,'flu-h-3'])/3)
```

```python
df.to_csv('datasets/annotation/Annotation_all.csv', index=False)
```
