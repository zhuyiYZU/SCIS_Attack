# Process Dataset

```python
import pandas as pd
```

```python
def dataset_creation_1(dataset_name, num, output_filename):
    dataset_1 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-one-token-fast-gpu.csv'
    dataset_2 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-two-token-fast-gpu.csv'
    dataset_3 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-three-token-fast-gpu.csv'
    dataset_4 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-four-token-fast-gpu.csv'
    dataset_5 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-five-token-fast-gpu.csv'
    dataset_6 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-random-token-fast-gpu.csv'
    dataset_7 = 'datasets/full-dataset-ensemble-len/' + dataset_name + '-4-sentence-fast-gpu.csv'
    
    df_1 = pd.read_csv(dataset_1)
    df_2 = pd.read_csv(dataset_2)
    df_3 = pd.read_csv(dataset_3)
    df_4 = pd.read_csv(dataset_4)
    df_5 = pd.read_csv(dataset_5)
    df_6 = pd.read_csv(dataset_6)
    df_7 = pd.read_csv(dataset_7)
    
    df = pd.DataFrame(columns=['index'])
    for i in range(0,num):
        # df.loc[i, 'Reference'] = df_1.loc[i, 'original']
        df.loc[i, 'a'] = df_1.loc[i, 'sampled']
        df.loc[i, 'coh-a'] = 0
        df.loc[i, 'flu-a'] = 0
        
        df.loc[i, 'b'] = df_1.loc[i, 'completed']
        df.loc[i, 'coh-b'] = 0
        df.loc[i, 'flu-b'] = 0
        
        df.loc[i, 'c'] = df_2.loc[i, 'completed']
        df.loc[i, 'coh-c'] = 0
        df.loc[i, 'flu-c'] = 0
        
        df.loc[i, 'd'] = df_3.loc[i, 'completed']
        df.loc[i, 'coh-d'] = 0
        df.loc[i, 'flu-d'] = 0
        
        df.loc[i, 'e'] = df_4.loc[i, 'completed']
        df.loc[i, 'coh-e'] = 0
        df.loc[i, 'flu-e'] = 0
        
        df.loc[i, 'f'] = df_5.loc[i, 'completed']
        df.loc[i, 'coh-f'] = 0
        df.loc[i, 'flu-f'] = 0
        
        df.loc[i, 'g'] = df_6.loc[i, 'completed']
        df.loc[i, 'coh-g'] = 0
        df.loc[i, 'flu-g'] = 0
        
        df.loc[i, 'h'] = df_7.loc[i, 'completed']
        df.loc[i, 'coh-h'] = 0
        df.loc[i, 'flu-h'] = 0
        
        df.loc[i, 'best'] = 'a'
        
    df.to_csv(output_filename, index=False)
```
